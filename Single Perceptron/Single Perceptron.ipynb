```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("horseRacing.csv")
dataset.head(10)
```

```python
#drop price column
dataset=dataset.drop(['price'],axis=1)
dataset.head(10)
```

```python
#position column replace 1 as 1 otherwise 0
dataset['calc_position']=np.where(dataset['calc_position']==1,1,0)
dataset.head(10)
```

```python
#calc_position visualization
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='calc_position',data=dataset,palette='hls')
plt.show()
    
```

```python
dataset=dataset.dropna(subset=dataset.iloc[:,2:-1].columns , how='all')
dataset.shape

    
```

```python
dataset.head(10)
```

```python
#replace missing values with mean for each column of the data
for i in dataset.iloc[:,2:-1].columns:
    dataset[i]=dataset[i].fillna(dataset[i].mean())

```

```python
dataset.head(10)
```

```python
dataset.isnull().sum()
```

```python
#MLP classifier
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
# apply mlp classifier
X=dataset.iloc[:,2:-1]
y=dataset.iloc[:,-1]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
mlp=MLPClassifier(hidden_layer_sizes=(10,10,10),max_iter=1000)
mlp.fit(X_train,y_train)
y_pred=mlp.predict(X_test)

#accuracy
print(accuracy_score(y_test,y_pred))
                                
```

```python
import pandas as pd
import numpy as np
```

```python
# input data
X = np.array([[0,0],[0,1],[1,0],[1,1]])

# output data
y = np.array([0,0,0,1])

# weights
w = np.array([0.0,0.0])
# bias
b = 0.0

# learning rate
learning_rate = 0.1

# activation function
def activation(x):
    if x > 0:
        return 1
    else:
        return 0
    
# prediction function
def predict(X):
    z = np.dot(X,w) + b
    a = activation(z)
    return a

# training function
def train(X,y,epochs, b):
    for i in range(epochs):
        for j in range(len(X)):
            a = predict(X[j])
            error = y[j] - a
            w[0] = w[0] + learning_rate * error * X[j][0]
            w[1] = w[1] + learning_rate * error * X[j][1]
            b = b + learning_rate * error
    print(w,b)
    
# training
train(X,y,20, b)
 
# prediction
for i in range(len(X)):
    print("X: {} y: {} prediction: {}".format(X[i],y[i],predict(X[i])))
```
