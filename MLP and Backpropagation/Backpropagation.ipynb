```python
import numpy as np
import pandas as pd
```

```python
dataset=pd.read_csv("horseRacing.csv")
dataset.head(10)
```

```python
#drop price column
dataset=dataset.drop(['price'],axis=1)
dataset.head(10)
```

```python
#position column replace 1 as 1 otherwise 0
dataset['calc_position']=np.where(dataset['calc_position']==1,1,0)
dataset.head(10)
```

```python
#calc_position visualization
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='calc_position',data=dataset,palette='hls')
plt.show()
    
```

```python
dataset=dataset.dropna(subset=dataset.iloc[:,2:-1].columns , how='all')
dataset.shape

    
```

```python
#replace missing values with mean for each column of the data
for i in dataset.iloc[:,2:-1].columns:
    dataset[i]=dataset[i].fillna(dataset[i].mean())

```

```python
dataset.head(10)
```

```python
#drop the horse_id
dataset=dataset.drop(['horse_id'],axis=1)
dataset.head(10)
```

```python
#drop the race_id column
dataset=dataset.drop(['race_id'],axis=1)
dataset.head(10)
```

```python

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# derivative of the sigmoid function
def sigmoid_derivative(x):
    return x * (1 - x)
```

```python
# the input and output
X=dataset.iloc[:,2:-1].values
y=dataset.iloc[:,-1].values
```

```python

#split the data into train and test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#define the number of input, hidden and output nodes
input_nodes = X_train.shape[1]
hidden_nodes = 3
output_nodes = 1

# learning rate
lr = 0.1

# define the number of samples
n_samples = X_train.shape[0]


# define the number of epochs
epochs = 100

```

```python
# initialize the weights
weights_input_hidden = np.random.normal(0, 1, (input_nodes, hidden_nodes))
weights_hidden_output = np.random.normal(0, 1, (hidden_nodes, output_nodes))

# initialize the bias
bias_hidden = np.random.normal(0, 1, (1, hidden_nodes))
bias_output = np.random.normal(0, 1, (1, output_nodes))
```

```python

# this is a stochastic approach, so we will use only one sample at a time
for i in range(n_samples):
    # forward propagation
    # calculate the input to the hidden layer
    hidden_layer_input = np.dot(X_train[i], weights_input_hidden) + bias_hidden
    hidden_layer_output = sigmoid(hidden_layer_input)
    
    # calculate the input to the output layer
    output_layer_input = np.dot(hidden_layer_output, weights_hidden_output) + bias_output
    output_layer_output = sigmoid(output_layer_input)

    # backpropagation
    # calculate the error at the output layer
    error_output = y_train[i] - output_layer_output
    error_hidden = np.dot(error_output, weights_hidden_output.T)
    
    # update the weights at the output layer
    weights_hidden_output += lr * np.dot(hidden_layer_output.T, error_output)
    weights_input_hidden += lr * np.dot(X_train[i].reshape(input_nodes, 1), error_hidden)

```
